```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#######散点图
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x,y)
plt.show()
```

```python
# 调整的散点图的参数
print help(plt.scatter)
```

```python
# s是面积，c是颜色，marker是形状，alpha是透明度
plt.scatter(x,y,s=100,c='y',marker='.',alpha=0.5)
plt.show()

# print help(np.loadtxt)
```

```python
#####折现图，随意从外部读取一个csv文件(using loadtxt in numpy)，然后画出对应的折线图
x,y = np.loadtxt('test2.csv',delimiter=',',skiprows=1,usecols=(6,7),unpack=True)
print x.shape
print y.shape
plt.plot(x,y)
plt.show()
```

```python
#####条形图
print help(plt.bar)
```

```python
#竖状的条形图
height_A = [10,20,5,30]
height_B = [50,20,12,4]
index = np.arange(4)
plt.bar(index,height_A,width=0.3,color='r',bottom=0)
# 层叠在原来条形图上
# plt.bar(index,height_B,width=0.3,bottom=height_A)
# 挨在原来条形图
plt.bar(index+0.3,height_B,color='y',width=0.3)
plt.show()
```

```python
# 横状的条形图,类似，注意width变为了height了
plt.barh(index,height_A,height=0.3)
plt.show()
```
