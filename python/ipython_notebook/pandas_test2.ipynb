```python
# 使用pandas分析外部文件
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
print help(pd.read_table)
```

```python
unames = ['user_id','gender','age','occupation','zip']
# 使用read_table读取数据
users = pd.read_table('data/users.dat',sep='::',header=None,names=unames)
users.head(10)
```

```python
ratnames = ['user_id','movie_id','rating','timestamp']
ratings = pd.read_table('data/ratings.dat',sep='::',header=None,names=ratnames)

movnames = ['movie_id','title','genres']
movies = pd.read_table('data/movies.dat',sep='::',header=None,names=movnames)

ratings.tail(5)
# movies.head(5)
```

```python
# 合并三张表,使用merge
data = pd.merge(pd.merge(users,ratings),movies)#使用了SQL语句来合并三张表
print data.shape
data.head(5)
```

```python
#找出用户id是2的所有用户的信息
data[data.user_id == 2].head(5)
```

```python
print help(data.pivot_table)
```

```python
# 找出每个电影的男性评分和女性评分
ratings_by_gender = data.pivot_table(values='rating',index='title',columns='gender',aggfunc='mean')
ratings_by_gender.head(5)#由于我减少了数据量，因此有NaN
```

```python
# 加入一个diff列
ratings_by_gender['diff'] = ratings_by_gender.F-ratings_by_gender.M
ratings_by_gender.head(5)
```

```python
# 将diff列的元素进行排序
ratings_by_gender.sort_values(by='diff',ascending=True).head(10)
```
